```python
# library & dataset
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('/content/heart_2020_cleaned.csv')
 
# left
sns.pairplot(df, kind="scatter", hue="AgeCategory", palette="Set2")
plt.show()
 
# right: you can give other arguments with plot_kws.
sns.pairplot(df, kind="scatter", hue="AgeCategory", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```
