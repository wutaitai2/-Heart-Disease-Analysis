```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path= os.path.join("data",'heart_2020_cleaned.csv' )
df = pd.read_csv(path)
df['Patient_ID'] = df.reset_index().index
df
```

```python
heartcount=df['HeartDisease'].groupby(df["Race"]).value_counts()
smokecount = df['Smoking'].groupby(df["Race"]).value_counts()
alccount = df['AlcoholDrinking'].groupby(df["Race"]).value_counts()
strokecount= df['Stroke'].groupby(df["Race"]).value_counts()
diabeticcount = df['Diabetic'].groupby(df["Race"]).value_counts()
asthmacount = df['Asthma'].groupby(df["Race"]).value_counts()
race = df["Race"].unique()
kidneyacount = df['KidneyDisease'].groupby(df["Race"]).value_counts()
skincount = df['SkinCancer'].groupby(df["Race"]).value_counts()
diffwalkingcount = df['DiffWalking'].groupby(df["Race"]).value_counts()
activitycount = df['PhysicalActivity'].groupby(df["Race"]).value_counts()
```

```python
RaceDisease = pd.concat([heartcount,smokecount, alccount , strokecount,diffwalkingcount, diabeticcount,activitycount,  asthmacount, kidneyacount, skincount ], axis=1)
RaceDisease.dropna()
```
